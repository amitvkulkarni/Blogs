## Example 1: Creating your first calendar heatmap
Let's begin by generating some random numbers for the year 2024 and then visualize the data on a calendar. Once the data is ready, simply call the calplot function to display the calendar heatmap.

```python
import calplot
import numpy as np; np.random.seed(sum(map(ord, 'calplot')))
import pandas as pd
all_days = pd.date_range('1/1/2024', periods=730, freq='D')
days = np.random.choice(all_days, 500)
events = pd.Series(np.random.randn(len(days)), index=days)
calplot.calplot(events)
```

## Example 2: Simulating a daily patient visit event
We can create a dataset that tracks daily patient visits to a healthcare facility, which is ideal for visualizing trends over time with a heatmap.

Objective:
* Generate a dataset for daily patient visits over the year 2024 (365 days).
* Add seasonal variations to simulate higher patient visits in seasonal  months (Jan, Feb, May, and June) and holidays (Dec)
* Visualize the data using a calendar heatmap with Calplot.

```python
# Generate dates for a year (365 days)
date_range_12months = pd.date_range(start="2024-01-01", end="2024-12-31", freq="D")

# Generate random patient visits using a Poisson distribution (most days will have around 50-150 visits)
patient_visits = np.random.poisson(lam=100, size=len(date_range_12months))

# Add seasonal trends (higher visits during flu season in winter and during holidays)
for month in [1, 2, 5, 6, 12]:  # Winter months (Jan, Feb) and December (holidays)
    patient_visits[date_range_12months.month == month] += np.random.randint(
        50, 200, size=len(date_range_12months[date_range_12months.month == month])
    )

# Create the DataFrame
patient_df = pd.DataFrame(
    {"Date": date_range_12months, "Patient Visits": patient_visits}
)

# Preview the data
patient_df.head(10)
```

```python
import calplot

# Prepare the data for Calplot
events = pd.Series(patient_df["Patient Visits"].values, index=patient_df["Date"])
```

```python
calplot.calplot(events, edgecolor=None, cmap="YlGn")
```

* Patient Visits: The data simulates daily patient visits using a Poisson distribution with added seasonal trends (change of seasons - winter & summer & holidays).
* Calplot Visualization: The heatmap generated using Calplot will help visualize the patient visits, showing patterns over time.

## Example 4: weekday vs weekend analysis
Healthcare centers often experience different patterns in patient visits on weekends compared to weekdays. This can provide valuable insights into how patient flow varies across different days. We will extend the code from the earlier section and add weekdays and weekend columns.

```python

# Add a column for day of the week (0=Monday, 6=Sunday)
patient_df["Day of Week"] = patient_df["Date"].dt.dayofweek

# Categorize weekends (Saturday=5, Sunday=6) and weekdays (Monday to Friday)
patient_df["Weekend"] = patient_df["Day of Week"].isin(
    [5, 6]
)  # True for weekends, False for weekdays

# Preview the data
patient_df.head(10)


```

* **Weekend Analysis**: The weekend column is used to filter weekends (Saturday and Sunday). We can now focus on the unique patterns observed during weekends.
* **Weekday Analysis**: Weekdays (Monday through Friday) are separated into their own analysis. This is useful to compare patient visits on regular weekdays versus weekends.

```python
# Separate data for weekends and weekdays
weekend_data = patient_df[patient_df["Weekend"] == True]
weekday_data = patient_df[patient_df["Weekend"] == False]

# Prepare the data for Calplot (weekends and weekdays)
events_weekend = pd.Series(
    weekend_data["Patient Visits"].values, index=weekend_data["Date"]
)
events_weekday = pd.Series(
    weekday_data["Patient Visits"].values, index=weekday_data["Date"]
)
```

```python
# Generate the calendar heatmap for weekends
calplot.calplot(events_weekend, yearlabel_kws={"color": "black"}, cmap="YlGn")
```

```python
# Generate the calendar heatmap for weekdays
calplot.calplot(events_weekday, yearlabel_kws={"color": "black"}, cmap="YlGn")
```

This method will help you better understand how patient visits differ between weekdays and weekends. This insight is invaluable for optimizing staffing, improving resource management, and gaining a deeper understanding of patient patterns.
