# **Context Sensitivity: Improving Performance of Language Models Using Flash Ranking**


In the fast-evolving world of artificial intelligence, the performance of language models is a hot topic. With applications ranging from chatbots to content creation, these models are transforming how we interact with technology. A critical aspect of their efficiency is how quickly and accurately they can rank and retrieve information. Here is a standard representation for building a RAG system where we load the documents, extract text from them, store it in a vector database, and retrieve the information based on the query.

This problem highlights the need for more effective ranking mechanisms that can pinpoint and elevate crucial data, regardless of its position within a text. Flash Ranking addresses this by ensuring that even the most deeply embedded information is efficiently retrieved and utilized. This blog will delve into the mechanics of Flash Ranking, exploring its benefits and its impact on the performance of language models. Here is a more advanced technique for developing an RAG system using FalshRank.

A crucial element of RAG's efficiency lies in its ability to retrieve information swiftly and accurately. This is where Flash Ranking becomes essential. Flash Ranking is a novel method that improves the speed and precision of information retrieval in language models, overcoming the limitations of traditional ranking systems that often struggle with resource consumption and sluggishness. A concept closely related to this discussion is the "Lost in the Middle" problem, where vital information embedded within long passages often goes unnoticed by conventional retrieval systems. ## Import the neccessary libraries

```python
import pandas as pd
import pickle
import faiss
import os
from langchain_huggingface import HuggingFaceEndpoint
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.chains import RetrievalQA
from langchain.document_loaders import PyPDFLoader
from langchain.document_loaders import DirectoryLoader, TextLoader
from langchain.vectorstores import FAISS
from InstructorEmbedding import INSTRUCTOR
from langchain.embeddings import HuggingFaceInstructEmbeddings
from langchain.prompts import ChatPromptTemplate, PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from flashrank.Ranker import Ranker, RerankRequest
import json
from langchain_chroma import Chroma
```

```python
from google.colab import userdata

userdata.get("HF_API_KEY")
userdata.get("HF_TOKEN")

print("API's loaded")
```

## Loading the data

```python
loader = DirectoryLoader(f"/content/data", glob="*.txt", loader_cls=TextLoader)
documents = loader.load()
documents
```

## Splitting the document using RecursiveCharacterTextSplitter

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=100, chunk_overlap=10)
texts = text_splitter.split_documents(documents)
texts
```

```python
# ! pip install sentence-transformers==2.2.2
```

```python
from langchain.embeddings import HuggingFaceInstructEmbeddings
instructor_embeddings = HuggingFaceInstructEmbeddings(model_name="hkunlp/instructor-xl")
```

## Storing the information in Vector database - FAISS

```python
import faiss
from langchain.vectorstores import FAISS
db_instructEmbedd = FAISS.from_documents(texts, instructor_embeddings)

retriever = db_instructEmbedd.as_retriever(search_kwargs={"k": 10})
retriever
```

## Testing the set up - Trail 1

```python
question = "Why are all the countries meeting and what is it about?"
doc = retriever.get_relevant_documents(question)
doc
```

## Formatting the content and converting it to dataframe

```python
formatted_data = []
for idx, doc in enumerate(doc):
    entry = {
        "id": idx + 1,  # Start id from 1
        "text": doc.page_content,
        # "meta": doc.metadata
    }
    formatted_data.append(entry)

# Convert to JSON string
formatted_data_json = json.dumps(formatted_data, indent=3)
# print(formatted_data_json)

# Parse the JSON data
parsed_data = json.loads(formatted_data_json)

# # Extract id and text into lists
ids = [item['id'] for item in parsed_data]
texts = [item['text'] for item in parsed_data]

# # Create a DataFrame
df_top_n_retrievals = pd.DataFrame({'id': ids, 'text': texts})
df_top_n_retrievals


```

## Using FlashRank

```python
from flashrank import Ranker, RerankRequest
ranker = Ranker()
```

```python
rerankrequest = RerankRequest(query=question, passages=formatted_data)
flashrank_results = ranker.rerank(rerankrequest)
flashrank_results
```

```python
df_re_rank = pd.DataFrame(flashrank_results)[['id', 'text']]
df_re_rank.rename(columns = {'id':'Re-rank-id', 'text': 'Re-rank-text'}, inplace = True)
df_re_rank
```

# Traditional approach Vs FlashRank approach

It is always interesting to examine the ranking of both approaches. Here are a few observations.
The traditional approach generated a response with 109 words and it answered our question. Most importantly the context was not lost and the response made sense.
The FashRank approach generated a shorter response with 83 words with context intact and a response much sharper.

```python
df_compare = pd.concat([df_top_n_retrievals, df_re_rank], axis = 1)
df_compare
```

# **Advance Your RAG System with Langchain’s Data Retrieval Strategies**
Insights into optimizing data retrieval efficiency and improving query relevance to boost overall system effectiveness with Long-Context Reorder, Cross Encoder Reranker, and LOTR

In the RAG system, retrieval entails retrieving pertinent information from a database by correlating keywords or contextual cues in user queries with stored documents, prioritizing those likely to offer precise answers. Achieving optimal relevance and accuracy in responses necessitates a comprehensive understanding of data characteristics and effective data processing techniques. Langchain streamlines the implementation of these techniques into concise code, facilitating enhanced data accessibility and refined retrieval outcomes.

# **Cross Encoder Reranker**

Cross Encoder Reranker is a tool that enhances search results ranking within RAG systems by assessing the semantic similarity and relevance of each document to the query. This method improves retrieval accuracy by prioritizing results that align with the query’s intent, thereby enhancing user satisfaction and system performance.

We will use distilbert from sentence transformers to create the embedding on the split documents, save it in the FAISS database, and configure it to fetch the top 10 most-suited documents based on our query. Note that we will use this response as a baseline to compare response from other techniques.

```python
from langchain_community.document_loaders import TextLoader
from langchain_community.vectorstores import FAISS
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
embeddingsModel = HuggingFaceEmbeddings(
    model_name="sentence-transformers/msmarco-distilbert-dot-v5"
)
retriever = FAISS.from_documents(texts, embeddingsModel).as_retriever(
    search_kwargs={"k": 10}
)

question = "Why are all the countries meeting and what is it about?"
docs = retriever.invoke(question)
docs
```

```python
df_distill = pd.DataFrame([(idx, doc.page_content) for idx, doc in enumerate(docs)])
# df_distill.rename(columns = {0: "Rank_distillbert",1: "text_distillbert"}, inplace = True)
df_distill.rename(columns = {0: "Rank_distillbert",1: "text"}, inplace = True)
df_distill
```

```python
from langchain.retrievers import ContextualCompressionRetriever
from langchain.retrievers.document_compressors import CrossEncoderReranker
from langchain_community.cross_encoders import HuggingFaceCrossEncoder
```

```python
model = HuggingFaceCrossEncoder(model_name="BAAI/bge-reranker-base")
compressor = CrossEncoderReranker(model=model, top_n=10)
compression_retriever = ContextualCompressionRetriever(
    base_compressor=compressor, base_retriever=retriever
)

compressed_docs = compression_retriever.invoke(question)
compressed_docs
```

```python
df_compressed = pd.DataFrame([(idx, doc.page_content) for idx, doc in enumerate(compressed_docs)])
# df_compressed.rename(columns = {0: "Rank_CrossEncoderReranker",1: "text_CrossEncoderReranker"}, inplace = True)
df_compressed.rename(columns = {0: "Rank_CrossEncoderReranker",1: "text"}, inplace = True)
df_compressed
```

```python
merged_df = pd.merge(df_distill, df_compressed, left_on='text', right_on='text')
merged_df
```

# **Long-Context Reorder**

Long-Context Reorder is a data retrieval technique that improves context understanding in RAG systems by prioritizing longer text or documents. This technique is particularly effective in scenarios where understanding the broader context of a query is crucial for accurate retrieval. It involves strategically reordering data sequences to optimize relevance and context coherence.

```python
from langchain_community.document_transformers import LongContextReorder
```

```python
reordering = LongContextReorder()
reordered_docs = reordering.transform_documents(docs)
df_longcontext_reorder = pd.DataFrame([(idx, doc.page_content) for idx, doc in enumerate(reordered_docs)])
df_longcontext_reorder.rename(columns = {0: "Rank_longcontext_reorder",1: "text"}, inplace = True)
df_longcontext_reorder
```

```python
merged_df = pd.merge(df_distill, df_longcontext_reorder, left_on='text', right_on='text')
merged_df
```

# **LOTR (Merger Retriever)**

LOTR, or Merger Retriever, is a technique that merges data from multiple sources into a unified output, enhancing efficiency in RAG systems. It minimizes redundancy by merging and deduplicating results from various processes, providing a consolidated view of information, and enabling quicker decision-making and deeper insights for users.

This will be slightly more complex than the previous two techniques. Let’s start by importing the necessary libraries.

```python
from langchain.retrievers.document_compressors import DocumentCompressorPipeline
from langchain.retrievers import (
    ContextualCompressionRetriever,
    MergerRetriever
)
```

```python
from langchain_community.document_transformers import (
    EmbeddingsClusteringFilter,
    EmbeddingsRedundantFilter,
)
```

```python
filter_embeddings  = HuggingFaceEmbeddings(
    model_name="sentence-transformers/msmarco-distilbert-dot-v5"
)
```

As the name suggests, it’s a merger retriever meaning there will be more than one model retrieving the documents. In our case, we will use two different embedding models namely all-MiniLM-L6-v2 and multi-qa-MiniLM-L6-dot-v1.

```python
model_all_mini = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
model_multi_qa_mini = HuggingFaceEmbeddings(model_name="multi-qa-MiniLM-L6-dot-v1")
```

The embeddings from each of the models are saved in the vector DB. We can use either the Chroma or FAISS but to make things more interesting, let’s use both. The first model will have Chroma and the second will have FAISS.

```python
retriever_all_mini_FAISS = FAISS.from_documents(texts, model_all_mini).as_retriever(
    search_kwargs={"k": 10}
)
```

```python
retriever_multi_qa_mini_Chroma = Chroma.from_documents(texts, model_all_mini).as_retriever(
    search_kwargs={"k": 10}
)
```

```python
lotr = MergerRetriever(retrievers=[retriever_all_mini_FAISS, retriever_multi_qa_mini_Chroma])
```

```python
q1 = lotr.get_relevant_documents(question)
q1
```

As the query and the number of documents to retrieve (k=10) are the same for both models, there will be duplicates in the retrieved documents and that doesn’t add any value and hence needs to be removed.

We will use a different embedding model to remove these duplicates which would also help us reduce any bias in the system. We can use any other model like OpenAI etc but we will use the model we had for the baseline version. In the end, we will check for the ranking of the retrieved documents.

```python
from langchain_community.document_transformers import LongContextReorder

filter = EmbeddingsRedundantFilter(embeddings=filter_embeddings)
reordering = LongContextReorder()
pipeline = DocumentCompressorPipeline(transformers=[filter, reordering])
compression_retriever_reordered = ContextualCompressionRetriever(
    base_compressor=pipeline, base_retriever=lotr
)
```

```python
df_lotr = pd.DataFrame([(idx, doc.page_content) for idx,doc in enumerate(compression_retriever_reordered.get_relevant_documents(question))], columns=['Rank_lotr', 'text'])
df_lotr
```

The ranking certainly is different from the baseline model and the other two models. Like in the previous two techniques, we will not be able to compare the results directly with the baseline model because we added a layer of additional embeddings on top of the merger retriever during the duplication removal phase.

For the sake of curiosity, let’s compare it with a baseline to check if we even retrieved any documents.

```python
merged_df = pd.merge(df_distill, df_lotr, left_on='text', right_on='text')
merged_df
```
